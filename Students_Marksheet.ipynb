```python
import pandas as pd 
import numpy  as np
import matplotlib.pyplot as plt 
```

# Load csv

```python
df = pd.read_csv("students_full_marksheet.csv")
```

```python
df
```

```python
df.head()
```

# Add a column Total 

```python
df['Total'] = df['Maths']+df['Physics']+df['Programming']+df['English']
```

```python
df
```

```python
df['Persentage'] = df['Total']/4
```

```python
df
```

# Add a column Grade (A, B, C based on total score)

```python
def grades(score):
    if score>=90:
        return '+A'
    elif score<90 and score>=80:
        return 'A'
    elif score<80 and score>=70:
        return 'B'
    elif score<70 and score>=60:
        return 'C'
    elif score<60 and score>=50:
        return 'D'
    else:
        return 'Fail'
```

```python
df['Grade'] = df['Persentage'].apply(grades)
```

```python
df
```

```python
df[df['Grade']=='A']
```

# Matplot lib

# Bar chart: Total marks of top 5 students

```python
fig , subplot1 = plt.subplots()
subplot1.bar(df_t5['Name'],df_t5['Total'])
subplot1.set(title='Top 5 Students', 
            xlabel= 'Names',
            ylabel= 'Marks')
plt.show()
```

# Line graph: Subject-wise average


```python
fig , subplot2= plt.subplots()
subplot2.plot(name,mean_scores)
subplot2.set(title='Averge Subject Scores', 
            xlabel= 'Subjects',
            ylabel= 'Average Scores')
plt.show()
```

# Pie chart: Grade distribution

```python
df['Grade'].value_counts()
```

```python
fig , subplot3= plt.subplots()
subplot3.pie(df['Grade'].value_counts(),
            labels=df['Grade'].value_counts().index, 
            startangle=90, 
             autopct='%1.1f%%')
subplot3.set(title='Grade Distribution')
plt.show()
```
